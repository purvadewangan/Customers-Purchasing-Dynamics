```python
#filling out the 37 nulls in review_rating with the category median 
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#renaming column
df.rename(columns ={"Purchase Amount (USD)": "purchase_amount"}, inplace=True)
```

```python
#identifying column names
df.columns
```

```python
#renaming columns
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
#creating a new column 'age_group'
labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]
df["age_group"]= pd.qcut(df["age"], q=4, labels = labels)
```

```python
df[['age', 'age_group']].tail(10)
```

```python
df.shape
```

```python
#finding unique items of 'frequency_of_purchases' column
df['frequency_of_purchases'].unique()
```

```python
#creating a column name purchasing_frequency
frequency_map = {"Weekly": 7,
                 "Fortnightly": 14,
                 "Bi-Weekly": 14,
                 "Monthly": 30,
                 "Every 3 Months": 90,
                 "Quarterly": 90,
                 "Annually": 365
                }
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_map)  
```

```python
df[["frequency_of_purchases", "purchase_frequency_days"]].head(10)
```
